# NA
NA is a logical constant of length 1 and is an indicator for a missing value.NA (capital letters) is a reserved word and can be coerced to any other data type vector (except raw) and can also be a product when importing data. NA and “NA” (as presented as string) are not interchangeable. NA stands for Not Available.

```r
typeof(NA) #type of NA in default is logical,but it can be coerced to any other data type vector
thisna <- NA
thisna == NA #if NA value compared by NA it will return NA too,so
is.na(thisna) #We can use "is.na" function for checking the NA value
vector <- c(1,4,6,4,NA)
typeof(NA_integer_) #Represented "integer"
typeof(NA_real_) #Represented "double"
typeof(NA_character_) #Represented "character"
typeof(NA_complex_) #Represented "complex"
lapply(vector,typeof) #NA can be coerced to any other data type depends on other data type in the vector (in this case it will be double)
```

```r
vector <- c(1,2,3,4,NA,NULL)
lists <- list(1,3,TRUE,NA,NULL)
vector #In vector,null has not shown
lists #But in list,null has shown
```

# NaN
NaN stands for Not A Number and is a logical vector of a length 1 and applies to numerical values, as well as real and imaginary parts of complex values, but not to values of integer vector. NaN is a reserved word.

```r
a <- 0/0 #Example
a
is.nan(NaN) #Function "is.nan" is for checking NaN value
is.na(NA) != is.nan(NA) #NA cant be checked using "is.nan" function
is.na(NaN) == is.nan(NaN) #NaN can be checked using "is.na" function
```

```r
#Read a csv file
dataset <- read.csv("https://academy.dqlab.id/dataset/dkikepadatankelurahan2013.csv",sep=",")
dataset
```

```r
#Show the columns name
names(dataset)
```

```r
#Changing column name
names(dataset)[8] <- "Not found"
names(dataset)
```

```r
#Changing columns name in a row
names(dataset)[c(8,9,10)] <- c("Not Found1","Not Found2","Not Found3")
names(dataset)
```

```r
#Deleting columns in dataset
dataset <- dataset[!names(dataset)%in%c('Not Found1','Not Found2','Not Found3','X.3','X.4','X.5','X.6','X.7','X.8','X.9','X.10','X.11')]
names(dataset)
```

```r
#Changing data type of column
dataset$NAMA.PROVINSI <- as.character(dataset$NAMA.PROVINSI)
str(dataset)
```

```r
#Taking column name with "grep" function
library(openxlsx)
dataset <- read.xlsx(xlsxFile="https://academy.dqlab.id/dataset/dkikepadatankelurahan2013.xlsx")
perempuan <- grep(pattern="perempuan",x=names(dataset),ignore.case=TRUE)
dataset[perempuan]
```

```r
#Adding new column with "rowSums" function
library(reshape2)
dataset$PEREMPUAN35KEATAS <- rowSums(dataset[perempuan])
dataset
```

```r
#Transform column to row
transformasi <- melt(data=dataset,id.vars=c("NAMA.KECAMATAN","NAMA.KELURAHAN"),measure.vars=c("35-39.Laki-Laki","35-39.Perempuan"),variable.name="Demo",value.name="JUMLAH")
transformasi
```

```r
#Splitting row using "colsplit"
transformasi[c("Umur","Jenis.Kelamin")] <- colsplit(transformasi$Demo,"\\.",c("Umur","Jenis.Kelamin"))
transformasi$Demo <- NULL
transformasi
```
