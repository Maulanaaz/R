```r
#Calling the dataset
library(openxlsx)
dataset <- read.xlsx(xlsxFile="https://academy.dqlab.id/dataset/dqlab_messy_data_pelanggan.xlsx",sheet="Pelanggan")
dataset
```

```r
#Profiling the dataset using "str"
str(dataset)
```

```r
#Profiling the dataset using "summary"
summary(dataset)
```

```r
#Changing the data types as they should be
dataset$Kode.Pelanggan <- as.factor(dataset$Kode.Pelanggan)
dataset$Nama.Lengkap <- as.factor(dataset$Nama.Lengkap)
dataset$Alamat <- as.factor(dataset$Alamat)
dataset$Tanggal.Lahir <- as.factor(dataset$Tanggal.Lahir)
dataset$Aktif <- as.factor(dataset$Aktif)
dataset$Kode.Pos <- as.factor(dataset$Kode.Pos)
dataset$No.Telepon <- as.factor(dataset$No.Telepon)
summary(dataset)
```

```r
#Profiling using Basic Pattern Analysis
library(bpa)
#Example
basic_pattern_analysis("DQLab")
basic_pattern_analysis("17 Agustus 1945")
basic_pattern_analysis("3.14")

#A-Z represented to A
#a-z represented to a
#0-9 represented to 9
#Symbols represented to Symbols
#Space represented to w
```

```r
#Profiling vector using BPA
vector <- c("I am","16","Years Old")
basic_pattern_analysis(vector)
```

```r
#Using "unique_only" parameter
vector <- c("Name","KD-0090","KD-0099")
basic_pattern_analysis(vector,unique_only=TRUE)
```

```r
#Profiling "Kode.Pelanggan" column using BPA
basic_pattern_analysis(dataset$Kode.Pelanggan,unique_only=TRUE)
```

```r
#Taking anomalous pattern from "Kode.Pelanggan"
basic_pattern_analysis(dataset$Kode.Pelanggan)=="AA-9999"
dataset$Kode.Pelanggan[basic_pattern_analysis(dataset$Kode.Pelanggan)=="AA-9999"]
```

```r
#Profiling "Nama.Lengkap" column
basic_pattern_analysis(dataset$Nama.Lengkap,unique_only=TRUE)
```

```r
#Introducing grepl as a regex in R
grepl(pattern="[A]","Customer") #is A exist in Customer?
grepl(pattern="[C]","Customer") #is C exist in Customer?
grepl(pattern="[Cr]","Customer") #is C or R exist in Customer?
grepl(pattern="[aa]","Customer") #are a(in a row) exist in Customer?
```

```r
#Finding anomalous patterns in "Nama.Lengkap"
dataset[grepl(pattern="[Aaw.,]",x=basic_pattern_analysis(dataset$Nama.Lengkap)),]
```

```r
#Profiling all columns
basic_pattern_analysis(dataset)
```

```r
#Combining normal dataset with its pattern
pattern <- basic_pattern_analysis(dataset)
names(pattern) <- paste("Pola",names(pattern),sep=".")
dataset <- cbind(dataset,pattern)
str(dataset)
```

```r
#Deleting double space on "Nama.Lengkap"
dataset$Nama.Lengkap <- gsub(" {2,}"," ",dataset$Nama.Lengkap)
dataset$Nama.Lengkap <- trimws(dataset$Nama.Lengkap,which="both")
dataset$Nama.Lengkap
```

```r
#Deleting number and symbols in "Nama.Lengkap"
dataset$Nama.Lengkap <- gsub("[^A-Za-z., ]","",dataset$Nama.Lengkap)
dataset$Nama.Lengkap
```

```r
#Standaritation "No.Telepon" 1
basic_pattern_analysis(dataset$No.Telepon,unique_only=TRUE)
dataset$No.Telepon <- as.character(dataset$No.Telepon)
dataset[dataset$Pola.No.Telepon=='9999999999999999',]$No.Telepon <- gsub("^6","+6",dataset[dataset$Pola.No.Telepon=='9999999999999999',]$No.Telepon)
dataset[dataset$Pola.No.Telepon=='9999999999999999',]$No.Telepon
```

```r
#Standaritation "No.Telepon" 2
dataset[dataset$Pola.No.Telepon=='999999999999999',]$No.Telepon <- gsub("^0","+62",dataset[dataset$Pola.No.Telepon=='999999999999999',]$No.Telepon)
basic_pattern_analysis(dataset$No.Telepon,unique_only=TRUE)
```

```r
#Standaritation "Kode.Pos"
dataset$Kode.Pos <- as.character(dataset$Kode.Pos)
dataset$Kode.Pos <- gsub("O","0",dataset$Kode.Pos)
dataset$Kode.Pos <- gsub("I","1",dataset$Kode.Pos)
basic_pattern_analysis(dataset$Kode.Pos,unique_only=TRUE)
```

```r
#Standaritation "Alamat" 1
grepl(pattern="^[A-Za-z]+$",x=dataset$Alamat)
grepl(pattern="^[aWA]+$",x=dataset$Pola.Alamat)

#Its mean there are no errors in the first technique
```

```r
#Standaritation "Alamat" 2
dataset$Alamat <- gsub("\\bjl\\b","Jalan",dataset$Alamat,ignore.case=TRUE)
dataset$Alamat <- gsub("jl [ ]*\\.","Jalan",dataset$Alamat,ignore.case=TRUE)
dataset$Alamat <- gsub("\\bjln\\b","Jalan",dataset$Alamat,ignore.case=TRUE)
dataset$Alamat <- gsub("jln [ ]*\\.","Jalan",dataset$Alamat,ignore.case=TRUE)
dataset$Alamat <- gsub("jalan\\.","Jalan",dataset$Alamat,ignore.case=TRUE)
dataset$Alamat
```

```r
#Standaritation "Aktif"
dataset$Aktif <- as.character(dataset$Aktif)
dataset$Aktif <- gsub("FALSE","0",dataset$Aktif)
dataset$Aktif <- gsub("TRUE","1",dataset$Aktif)
dataset$Aktif <- gsub("O","0",dataset$Aktif)
dataset$Aktif <- gsub("I","1",dataset$Aktif)
basic_pattern_analysis(dataset$Aktif,unique_only=TRUE)
```

```r
#Standaritation ("Tanggal.Lahir")
dataset$Tanggal.Lahir <- as.character(dataset$Tanggal.Lahir)
dataset[grepl(pattern="[Aa]",x=basic_pattern_analysis(dataset$Tanggal.Lahir)),]$Tanggal.Lahir
dataset$Tanggal.Lahir <- gsub("Januari","-01-",dataset$Tanggal.Lahir)
dataset$Tanggal.Lahir <- gsub("Februari","-02-",dataset$Tanggal.Lahir)
dataset$Tanggal.Lahir <- gsub("Maret","-03-",dataset$Tanggal.Lahir)
dataset$Tanggal.Lahir <- gsub("April","-04-",dataset$Tanggal.Lahir)
dataset$Tanggal.Lahir <- gsub("Mei","-05-",dataset$Tanggal.Lahir)
dataset$Tanggal.Lahir <- gsub("Juni","-06-",dataset$Tanggal.Lahir)
dataset$Tanggal.Lahir <- gsub("Juli","-07-",dataset$Tanggal.Lahir)
dataset$Tanggal.Lahir <- gsub("Agustus","-08-",dataset$Tanggal.Lahir)
dataset$Tanggal.Lahir <- gsub("September","-09-",dataset$Tanggal.Lahir)
dataset$Tanggal.Lahir <- gsub("Oktober","-10-",dataset$Tanggal.Lahir)
dataset$Tanggal.Lahir <- gsub("November","-11-",dataset$Tanggal.Lahir)
dataset$Tanggal.Lahir <- gsub("Desember","-12-",dataset$Tanggal.Lahir)
unique(dataset$Tanggal.Lahir)
```

```r
#Standaritation "Tanggal.Lahir" 2
library(reshape2)
tanggal.split <- colsplit(dataset[grepl(pattern="[/]",x=basic_pattern_analysis(dataset$Tanggal.Lahir)),]$Tanggal.Lahir,"/",c('Bulan','Hari','Tahun'))
ab <- dataset[grepl(pattern="[/]",x=basic_pattern_analysis(dataset$Tanggal.Lahir)),]$Tanggal.Lahir
tanggal.split$Tahun <- sapply(tanggal.split$Tahun, function(x) if (x>=0 & x<10) 2000+x else if (x>=10 & x<100) 1900+x else x)
ab <- paste(tanggal.split$Hari,tanggal.split$Bulan,tanggal.split$Tahun, sep = "-")
ab
ac <- dataset[!grepl(pattern="[/]",x=basic_pattern_analysis(dataset$Tanggal.Lahir)),]$Tanggal.Lahir
var1 <- dataset[!grepl(pattern="[/]",x=basic_pattern_analysis(dataset$Tanggal.Lahir)),]$Kode.Pelanggan
var2 <- dataset[grepl(pattern="[/]",x=basic_pattern_analysis(dataset$Tanggal.Lahir)),]$Kode.Pelanggan
data1 <- data.frame("Kode.Pelanggan" = var1,"Tanggal.Lahir" = ac)
data2 <- data.frame("Kode.Pelanggan" = var2,"Tanggal.Lahir" = ab)
data.gabungan <- rbind(data1,data2)
data.gabungan
```

```r
#Standaritation "Tanggal.Lahir" 3
data.gabungan$Tanggal.Lahir <- format(as.Date(data.gabungan$Tanggal.Lahir,'%d-%m-%Y'),'%d-%m-%Y')
data.gabungan
```
