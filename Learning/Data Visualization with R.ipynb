The ggplot2 package, created by Hadley Wickham, offers a powerful graphics language for creating elegant and complex plots. Its popularity in the R community has exploded in recent years. Originally based on Leland Wilkinson's The Grammar of Graphics, ggplot2 allows you to create graphs that represent both univariate and multivariate numerical and categorical data in a straightforward manner. Grouping can be represented by color, symbol, size, and transparency. The creation of trellis plots (i.e., conditioning) is relatively simple.

```r
#Creating a blank canva
library(ggplot2)
ggplot()
```

```r
#Adding title in the graph
ggplot() + labs(title="This is title")
```

```r
#Storing the plot to the variable
plot <- ggplot()
plot <- plot + labs(title="Blank Canva")
plot
```

```r
#Labelling axises and adding subtitle
plot <- ggplot()
plot <- plot + labs(title="Grafik Penduduk",subtitle="Bagian-1")
plot <- plot + labs(x="Nama Kabupaten",y="Jumlah Penduduk")
plot
```

```r
#Using summary function on the graph object
summary(plot)
```

```r
#Reading the data that will be used to making graph
dataset <- read.csv("https://academy.dqlab.id/dataset/dkikepadatankelurahan2013.csv",sep=",")
dataset
```

```r
#Mapping x,y and color using "aes" function
plot <- ggplot(data=dataset,aes(x=LUAS.WILAYAH..KM2.,y=KEPADATAN..JIWA.KM2.,color=NAMA.KABUPATEN.KOTA))
summary(plot)
```

```r
#Displaying the plot
plot
```

```r
#Creating scatter plot using layer
dataset <- read.csv("https://academy.dqlab.id/dataset/dkikepadatankelurahan2013.csv",sep=",")
plot <- ggplot(data=dataset,aes(x=LUAS.WILAYAH..KM2.,y=KEPADATAN..JIWA.KM2.,color=NAMA.KABUPATEN.KOTA))
plot <- plot + layer(geom="point",stat="identity",position="identity")
plot
```

```r
#Creating scatter plot using geom_point()
dataset <- read.csv("https://academy.dqlab.id/dataset/dkikepadatankelurahan2013.csv",sep=",")
plot <- ggplot(data=dataset,aes(x=LUAS.WILAYAH..KM2.,y=KEPADATAN..JIWA.KM2.,color=NAMA.KABUPATEN.KOTA))
plot <- plot + geom_point()
plot
```

```r
#Adding title and labelling
dataset <- read.csv("https://academy.dqlab.id/dataset/dkikepadatankelurahan2013.csv",sep=",")
plot <- ggplot(data=dataset,aes(x=LUAS.WILAYAH..KM2.,y=KEPADATAN..JIWA.KM2.,color=NAMA.KABUPATEN.KOTA))
plot <- plot + theme(plot.title=element_text(hjust=0.5))
plot <- plot + labs(title="Luas Wilayah vs Kepadatan Penduduk",x="Luas Wilayah (km2)",y="Kepadatan Penduduk (km2)",color="Nama Kabupaten/Kota")
plot <- plot + geom_point()
plot
```

```r
#Creating histogram plot using geom_histogram()
dataset <- read.csv("https://academy.dqlab.id/dataset/dkikepadatankelurahan2013.csv",sep=",")
plot <- ggplot(data=dataset,aes(x=KEPADATAN..JIWA.KM2.))
plot <- plot + geom_histogram(binwidth=10000)
plot
```

```r
#Adding fill aesthetic on the graph
dataset <- read.csv("https://academy.dqlab.id/dataset/dkikepadatankelurahan2013.csv",sep=",")
plot <- ggplot(data=dataset,aes(x=KEPADATAN..JIWA.KM2.,fill=NAMA.KABUPATEN.KOTA))
plot <- plot + geom_histogram(binwidth=10000)
plot
```

```r
#Creating bar graph using geom_bar
dataset <- read.csv("https://academy.dqlab.id/dataset/datakependudukandki-dqlab.csv",sep=",")
plot <- ggplot(data=dataset,aes(x=NAMA.KABUPATEN.KOTA))
plot <- plot + geom_bar()
plot
```

```r
#Creating bar graph with y aesthetic and stat identity
dataset <- read.csv("https://academy.dqlab.id/dataset/datakependudukandki-dqlab.csv",sep=",")
plot <- ggplot(data=dataset,aes(x=NAMA.KABUPATEN.KOTA,y=JUMLAH))
plot <- plot + geom_bar(stat="identity")
plot
```

```r
#Creating bar graph with fill aesthetic and dodge position
dataset <- read.csv("https://academy.dqlab.id/dataset/datakependudukandki-dqlab.csv",sep=",")
plot <- ggplot(data=dataset,aes(x=NAMA.KABUPATEN.KOTA,y=JUMLAH,fill=JENIS.KELAMIN))
plot <- plot + geom_bar(stat="identity",position="dodge")
plot
```

```r
#Using aggregate function for finding mean
p <- aggregate(x=list(RATA.RATA = dataset$JUMLAH),FUN="mean",by=list(NAMA.KABUPATEN.KOTA=dataset$NAMA.KABUPATEN.KOTA,JENIS.KELAMIN=dataset$JENIS.KELAMIN))
p
```

```r
#Tidying up the graph
dataset <- read.csv("https://academy.dqlab.id/dataset/datakependudukandki-dqlab.csv",sep=",")
p <- aggregate(x=list(JUMLAH=dataset$JUMLAH),FUN="sum",by=list(NAMA.KABUPATEN.KOTA=dataset$NAMA.KABUPATEN.KOTA,JENIS.KELAMIN=dataset$JENIS.KELAMIN))
plot <- ggplot(data=p,aes(x=NAMA.KABUPATEN.KOTA,y=JUMLAH,fill=JENIS.KELAMIN,label=JUMLAH))
plot <- plot + geom_bar(stat="identity",position="dodge")
plot <- plot + labs(title="JUMLAH PENDUDUK > 35 TAHUN",x="Nama Kabupaten/Kota",y="Jumlah",fill="Jenis Kelamin")
plot <- plot + theme(axis.text.x=element_text(angle=45,vjust=0.5),plot.title=element_text(hjust=0.5))
plot <- plot + geom_text(position=position_dodge(1.2))
plot
```

```r
#Faceting on scatter plot
dataset <- read.csv("https://academy.dqlab.id/dataset/dkikepadatankelurahan2013.csv",sep=",")
plot <- ggplot(data=dataset,aes(x=LUAS.WILAYAH..KM2.,y=KEPADATAN..JIWA.KM2.,color=NAMA.KABUPATEN.KOTA))
plot <- plot + theme(plot.title=element_text(hjust=0.5))
plot <- plot + labs(title="Luas Wilayah vs Kepadatan Penduduk",x="Luas Wilayah (km2)",y="Kepadatan Penduduk (km2)",color="Nama Kabupaten/Kota")
plot <- plot + geom_point()
plot <- plot + facet_wrap(~NAMA.KABUPATEN.KOTA,ncol=2)
plot
```

```r
#Faceting on Histogram plot
dataset <- read.csv("https://academy.dqlab.id/dataset/dkikepadatankelurahan2013.csv",sep=",")
plot <- ggplot(data=dataset,aes(x=KEPADATAN..JIWA.KM2.,fill=NAMA.KABUPATEN.KOTA))
plot <- plot + geom_histogram(binwidth=10000)
plot <- plot + facet_wrap(~NAMA.KABUPATEN.KOTA,ncol=2)
plot
```
