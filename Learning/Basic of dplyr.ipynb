```r
library(dplyr)
library(hflights)
dataset <- tbl_df(hflights)
dataset
```

### Filter Function

```r
filtered <- dataset[dataset$Month==1 & dataset$DayOfWeek==2,] #Normal statement
head(filtered)
filtered2 <- filter(dataset,Month==1,DayOfWeek==2) #Use filter function
head(filtered2)
```

```r
#OR operator
nofilter <- dataset[dataset$DayOfWeek==2 | dataset$DayOfWeek==7,] #Normal
head(nofilter)
filtered <- filter(dataset,DayOfWeek==2|DayOfWeek==7)#With filter function
#or
filtered <- filter(dataset,DayOfWeek%in%c(2,7))
head(filtered)
```

### Select Function

```r
head(dataset[,c("Month","DepTime","ArrTime")]) #Normal Statement
head(select(dataset,Month,DepTime,ArrTime)) #Use Select Function
```

```r
head(dataset[,c("Year","DayofMonth","DepTime","ArrTime","DepDelay","ArrDelay")]) #Normal
head(select(dataset,Year:DayofMonth,contains("Dep"),contains("Delay"))) #Use select function
```

### Chaining Method

```r
head(filter(select(dataset,Year:DepTime,contains("Delay")),DayofMonth==1)) #Without chaining
dataset %>% #With Chaining
select(Year:DepTime,contains("Delay")) %>%
filter(DayofMonth==1)
```

```r
x <- 1:10 ; y <- 11:20
sqrt(sum((x+y)^2)) # without chaining
(x+y)^2 %>% sum() %>% sqrt() #with chaining
```

```r
tabel <- read.csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/lo5_m01_mp01.csv")
  tabel %>% #with chaining
  mutate(Laki.Laki = -Laki.laki) %>%
  arrange(desc(Kelompok.Usia))
arrange(mutate(tabel,Laki.Laki=-Laki.laki),desc(Kelompok.Usia)) #without chaining
```

### Arrange function

```r
dataset[order(dataset$ArrDelay),c("UniqueCarrier","ArrDelay")] #without arrange function
dataset %>% #with chaining method and arrange function(descending)
select(UniqueCarrier,ArrDelay) %>%
arrange(desc(ArrDelay))
```

### Mutate Function

```r
dataset$Speed <- (dataset$Distance/dataset$ActualElapsedTime) #without mutate function
dataset[,c("ActualElapsedTime","Distance","Speed")]
dataset$Speed <- NULL
dataset %>% #with mutate function
select(ActualElapsedTime,Distance) %>%
mutate(Speed=Distance/ActualElapsedTime)
```
